```python
def getLights(filename):
    f = open(filename)
    lights = []
    for l in f.readlines():
        row = []
        for p in l.strip("\n"):
            if p=="#":
                row.append(1)
            else:
                row.append(0)
        lights.append(row)
    return lights
```

```python
#lights = getLights("./data/day18test0.txt")
lights = getLights("./data/input18.txt")

import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,10))
f = plt.imshow(lights,cmap='binary')
f.axes.get_xaxis().set_visible(False)
f.axes.get_yaxis().set_visible(False)
```

```python
def getNeighbors(p,lights):
    r0,c0 = p
    return [ (r,c) for (r,c) in [(r0-1, c0-1), (r0-1, c0), (r0-1, c0+1), 
                                 (r0  , c0-1),             (r0  , c0+1),
                                 (r0+1, c0-1), (r0+1, c0), (r0+1, c0+1)] \
            if 0<=r<len(lights) and 0<=c<len(lights[0]) ]
p = (0,0)
getNeighbors(p,lights)
```

```python
from copy import deepcopy

#lights = getLights("./data/day18test0.txt")
#ni = 4

lights = getLights("./data/input18.txt")
ni = 100

for i in range(ni):
    
    newlights = deepcopy(lights)

    for r in range(len(lights)):
        for(c) in range(len(lights[0])):
            count = 0
            for rn,cn in getNeighbors((r,c),lights):
                count += lights[rn][cn]
            if lights[r][c]==1:
                if count==2 or count==3 :
                    newlights[r][c] = 1
                else:
                    newlights[r][c] = 0
            else:
                if count==3 :
                    newlights[r][c] = 1
                else:
                    newlights[r][c] = 0

    lights = deepcopy(newlights)

import numpy as np
lights = np.array(lights)
print(sum(sum(lights)))

import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,10))
plt.imshow(lights,cmap='binary')
f.axes.get_xaxis().set_visible(False)
f.axes.get_yaxis().set_visible(False)
```
